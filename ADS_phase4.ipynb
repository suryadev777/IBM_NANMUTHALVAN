```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense
```

```python
data = pd.read_csv("/kaggle/input/microsoft-lifetime-stocks-dataset/MSFT.csv")
```

```python
data.columns
```

# pre processing

```python
#Remove the Date column because it doesn't have any significance

data = data.drop(["Date"] ,axis=1)
```

```python
# checking for null values
null_counts = data.isnull()
null_counts.sum()
```

normalize data 

```python
scaler = MinMaxScaler(feature_range =(0,1))
dataset = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
dataset
```

**TRAIN TEST SPLIT**

```python
X = dataset.drop(["High"] , axis = 1)
Y = dataset["High"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y)
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(X_train,y_train,epochs=20,batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
1-mean_squared_error(y_test,y_pred)
```
