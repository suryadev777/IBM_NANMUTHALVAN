```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/kaggle/input/microsoft-lifetime-stocks-dataset/MSFT.csv")
```

```python
data.columns
```

# pre processing

```python
#Remove the Date column because it doesn't have any significance

data = data.drop(["Date"] ,axis=1)
```

```python
# checking for null values
null_counts = data.isnull()
null_counts.sum()
```

normalize data 

```python
scaler = MinMaxScaler(feature_range =(0,1))
dataset = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
dataset
```

**TRAIN TEST SPLIT**

```python
X = dataset.drop(["High"] , axis = 1)
Y = dataset["High"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y)
```
